```python
import torch
from diffusers import StableDiffusionPipeline, LMSDiscreteScheduler
from PIL import Image
```

```python
def load_model():
    try:
        # Load the pre-trained Stable Diffusion model from Hugging Face
        model_id = "CompVis/stable-diffusion-v1-4"
        device = "cuda" if torch.cuda.is_available() else "cpu"

        # Initialize the pipeline with the model and scheduler
        pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float32)
        pipe.scheduler = LMSDiscreteScheduler.from_config(pipe.scheduler.config)
        pipe = pipe.to(device)
        return pipe
    except Exception as e:
        print(f"Error loading model: {e}")
        return None
```

```python
def generate_image(pipe, prompt, output_file="output.png", guidance_scale=7.5, num_inference_steps=50, height=512, width=512):
    try:
        if pipe is None:
            raise ValueError("Model pipeline is not loaded correctly.")

        # Generate image
        image = pipe(prompt, guidance_scale=guidance_scale, num_inference_steps=num_inference_steps, height=height, width=width).images[0]

        # Save the image
        image.save(output_file)
        print(f"Image saved to {output_file}")
        return image
    except Exception as e:
        print(f"Error generating image: {e}")
        return None
```

```python
def main():
    # Load the model
    pipe = load_model()

    # Define your text prompt
    prompt = "A futuristic cityscape at sunset with flying cars and tall skyscrapers."

    # Generate and save the image
    generate_image(pipe, prompt)
```

```python
if __name__ == "__main__":
    main()
```
